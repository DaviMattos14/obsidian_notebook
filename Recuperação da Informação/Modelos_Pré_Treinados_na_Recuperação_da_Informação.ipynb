# Pré-processamento

## Configurando o Stemmer e a

```python
# Garantindo que as dependências NLTK estão baixadas
nltk.download('stopwords')
nltk.download('rslp')
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
stemmer = RSLPStemmer()
stopwords_pt = set(stopwords.words('portuguese'))

def clean_html(text):
    if isinstance(text, str):
        return re.sub(r'<[^>]+>', ' ', text) # Substitui tags por espaço
    return ''

def preprocess(text):
    if not isinstance(text, str):
      return []
    # Normaliza
    text = unidecode(text.lower())
    # Tokeniza e remove stopwords
    tokens = nltk.word_tokenize(text, language='portuguese')

    tokens_ = [
        stemmer.stem(palavra)
        for palavra in tokens
        if palavra.isalpha()  # remove números e pontuação
        and palavra not in stopwords_pt # Remove stopwords
        and len(palavra) > 2      # Remove palavras muito curtas
    ]
    return tokens_
```

# Modelo BM25

## Quati


### Montando o BM25

```python
from rank_bm25 import BM25Okapi
from tqdm.auto import tqdm

corpus_quati_tokens = quati_docs_sample['text'].tolist()

quati_doc_ids_map = quati_docs_sample['doc_id'].tolist()

bm25_quati_amostra = BM25Okapi(corpus_quati_tokens)
```

### Passando as Consultas

```python
run_quati = {}
top_k = 100 # Vamos recuperar os 100 melhores documentos para avaliação

for index, row in tqdm(quati_query_sample.iterrows(), total=len(quati_query_sample)):
    query_id = row['qid']
    query_tokens = row['query'] # Já está pré-processada

    # 1. Pega os scores do BM25 para esta consulta
    doc_scores = bm25_quati_amostra.get_scores(query_tokens)

    # 2. Combina os IDs dos documentos (do mapa) com seus scores
    scores_com_ids = zip(quati_doc_ids_map, doc_scores)

    # 3. Ordena por score (decrescente) e pega os top_k
    top_results = sorted(scores_com_ids, key=lambda x: x[1], reverse=True)[:top_k]

    # 4. Salva no formato do 'run' (apenas docs com score > 0)
    run_quati[query_id] = {doc_id: score for doc_id, score in top_results if score > 0}

print("Buscas do Quati concluídas!")
```

### Avaliação

```python
import ranx

# 1. Formatar os Qrels (gabarito) para o formato dict
qrels_quati_dict = {}
for index, row in quati_qrels_sample.iterrows():
    qid = row['qid']
    doc_id = row['doc_id']
    label = row['score']

    if qid not in qrels_quati_dict:
        qrels_quati_dict[qid] = {}

    # Adicionamos apenas os relevantes (score > 0)
    if label > 0:
        qrels_quati_dict[qid][doc_id] = label

# 2. Converter para os objetos do ranx
qrels_quati = ranx.Qrels(qrels_quati_dict)
run_quati_obj = ranx.Run(run_quati)


metrics = ["ndcg@10", "map@10", "recall@100", "precision@10", "f1@10", "r-precision"]

# 4. Avaliar!
report_quati = ranx.evaluate(
    qrels=qrels_quati,
    run=run_quati_obj,
    metrics=metrics,
    make_comparable=True
)

print("\n--- 📊 Relatório de Avaliação (Quati Amostra) ---")
pd.DataFrame.from_dict(report_quati, orient='index')
```

## JurisTCU

### Montando o BM25

```python
corpus_juris_tokens = juris_docs_sample['tokens'].tolist()

juris_doc_ids_map = juris_docs_sample['doc_id'].tolist()

print(f"Iniciando indexação BM25 do JurisTCU ({len(corpus_juris_tokens)} docs)...")
bm25_juris = BM25Okapi(corpus_juris_tokens)
print("Indexação do JurisTCU concluída.")
```

### Passando as Consultas

```python
run_juris = {}
top_k = 100

print(f"Executando buscas para {len(juris_queries_sample)} consultas JurisTCU...")

for index, row in tqdm(juris_queries_sample.iterrows(), total=len(juris_queries_sample)):
    query_id = row['qid']
    query_tokens = row['query']

    doc_scores = bm25_juris.get_scores(query_tokens)
    scores_com_ids = zip(juris_doc_ids_map, doc_scores)
    top_results = sorted(scores_com_ids, key=lambda x: x[1], reverse=True)[:top_k]

    run_juris[query_id] = {str(doc_id): score for doc_id, score in top_results if score > 0}

print("Buscas do JurisTCU concluídas!")
```

## Curva de interpolação

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_precision_recall_curve(run_dict, qrels_dict, dataset_name="Dataset"):
    """
    Gera a curva de precisão interpolada (11 pontos padrão de revocação)
    a partir dos dicionários run e qrels.
    """
    recall_points = np.linspace(0, 1, 11)
    precisions = []

    for recall_level in recall_points:
        precisions_nivel = []
        for qid, retrieved_docs in run_dict.items():
            relevant_docs = qrels_dict.get(qid, {})
            total_relevant = len(relevant_docs)
            if total_relevant == 0:
                continue

            retrieved = list(retrieved_docs.keys())
            relevant_retrieved = [doc for doc in retrieved if doc in relevant_docs]

            precisao = []
            revocacao = []
            relevant_found = 0

            for i, doc in enumerate(retrieved):
                if doc in relevant_docs:
                    relevant_found += 1
                prec = relevant_found / (i + 1)
                rec = relevant_found / total_relevant
                precisao.append(prec)
                revocacao.append(rec)

            # Interpolação no ponto atual de recall
            prec_interp = max([p for p, r in zip(precisao, revocacao) if r >= recall_level], default=0)
            precisions_nivel.append(prec_interp)

        precisions.append(np.mean(precisions_nivel) if precisions_nivel else 0)

    plt.figure(figsize=(7,5))
    plt.plot(recall_points, precisions, marker='o', color='blue')
    plt.title(f"Curva de Precisão Interpolada — {dataset_name}")
    plt.xlabel("Revocação (Recall)")
    plt.ylabel("Precisão Interpolada")
    plt.grid(True)
    plt.show()
```
