```r
#2-modelando o lançamento de uma moeda 10 vezes
x<-sample(x=c(0,1),size=10,replace=T,prob=c(1/2,1/2))
x
```

```r
#3-contando o número de vezes com resultado igual a 1
x<-sample(x=c(0,1),size=10,replace=T,prob=c(1/2,1/2))
y<-sum(x)
y
```

```r
#4- repetindo o experimento 100 vezes e analisando resultados
N<-100
y<-vector(mode='numeric', length = N)
for (i in 1:N){
  x<-sample(x=c(0,1),size=10,replace=T,prob=c(1/2,1/2))
  y[i]<-sum(x)
}
#analisando resultados
summary(y)
hist(y)
mean(y)
var(y)
sd(y)


```

```r
#10-gerando cenários usando estimativa de três pontos: (Minimo,Máximo, Mais provável)

#Estimativa de tres pontos para uma obra (R$mil)
# Escv<-(75,92.5,82.5)
# Fund<-(57.5,77.5,67.5)
# Estr<-(430,472.5,445)
# Telh<-(140,157.5,145)
# Acab<-(72.5,107.5,92.5)


library (triangle)
Ns<-100
Ni<-5
rs<-vector(length=Ns)
pt=matrix(data=c( 75,92.5,82.5,
                  57.5,77.5,67.5,
                  430,472.5,445,
                  140,157.5,145,
                  72.5,107.5,92.5),
            nrow=Ni,ncol=3,byrow=TRUE)
cenarios<-matrix(ncol=Ni,nrow=Ns)

for (i in 1:Ns){
#gera um scenario
  for (j in 1:Ni){
    cenarios[i,j]<-rtriangle(1,pt[j,1],pt[j,2],pt[j,3])
  }
}
head(cenarios)
rs<-apply(cenarios,M=1,sum)
hist(rs)
plot(ecdf(rs))
rs
```

```r
#11-gerando cenarios 2: mais eficiente computacionalmente

construcao<-function(){
library (triangle)
#Constantes
Ns=1000

#Passo 1-Gerar as amostras das variaveis de risco
Escv<-rtriangle(Ns,75,92.5,82.5)
Fund<-rtriangle(Ns,57.5,77.5,67.5)
Estr<-rtriangle(Ns,430,472.5,445)
Telh<-rtriangle(Ns,140,157.5,145)
Acab<-rtriangle(Ns,72.5,107.5,92.5)
#Passo 2-Montar a matriz de cenarios
Cenarios<-cbind(Escv,Fund,Estr,Telh,Acab)

#Passo 3-Avaliar cenarios
Custo<-apply(Cenarios,M=1,sum)

#Passo 4-Mostrar resultados
hist(Custo)
plot(ecdf(Custo))

#Passo 5-Retorna resultado
Cenarios
}
construcao()
```
